```python
df = pd.read_csv("enrollment_data.csv")
df
```

```python
plt.xlabel('Years')
plt.ylabel('Count Of Enrollment')
plt.title('Student Enrollment Data')

plt.plot(df['Programming'], df['Digital Marketing'])
```

```python
plt.xlabel('Years')
plt.ylabel('Count Of Enrollment')
plt.title('Student Enrollment Data')

plt.plot(df['Year'], df['Programming'])
plt.plot(df['Year'], df['Digital Marketing'])
```

```python
plt.xlabel('Years')
plt.ylabel('Count Of Enrollment')
plt.title('Student Enrollment Data')

plt.plot(df['Year'], df['Programming'], label= 'Programming', color='red', linewidth=3,linestyle='dashdot', marker='o')
plt.plot(df['Year'], df['Digital Marketing'],  label= 'Digitial Marketing', color='green', linewidth=3,linestyle='dotted')

plt.legend()
plt.grid()
```
