```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
sns.set(font_scale=1.0)
# from google.colab import files
```

```python
df_titanic = pd.read_csv('D:/ML-AI/Phitron/Codebase/Semester-2/Codes/Assignment-2/practices/Titanic-Dataset.csv')
df_titanic.head()
```

```python
print("Dataset Shape:", df_titanic.shape)
df_titanic.info()
print("\nStatistical Summary:\n", df_titanic.describe(include='all').T)
print("\nMissing Values:\n", df_titanic.isnull().sum())
```

```python
print("Distribution of Age Column:")

plt.figure(figsize=(8, 6))
sns.histplot(df_titanic['Age'], bins=20, kde=True,)
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
print("Distribution of Embarked Column:")
plt.figure(figsize=(8, 6))
sns.countplot(data=df_titanic, x='Embarked', palette='Set1' )
plt.title('Embarked Distribution of Titanic Passengers')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.show()
```

```python
age_median = df_titanic['Age'].median()
df_titanic['Age'] = df_titanic['Age'].fillna(age_median)

embarked_mode = df_titanic['Embarked'].mode()[0]
df_titanic['Embarked'] = df_titanic['Embarked'].fillna(embarked_mode)

df_titanic = df_titanic.drop(columns=['Cabin'])
```
